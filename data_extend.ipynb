## EXTEND TABLE BASED ON A COLUMN WITH NEW VALUES FROM SEVERAL TEXT FILES

```python
df = pd.read_csv("/content/COF_CRAFTED_uptake_geometric.csv")
df.head()
```

```python

# Get a list of files in the separate folder
folder_path = "/content/drive/MyDrive/UoM Dissertation/COF_SURFACE"  # Replace with your folder path
files = os.listdir(folder_path)
#/content/drive/MyDrive/UoM Dissertation/all_surface_feature
```

```python

# Function to extract MOF name from filename or cif_files column
def get_mof_name(name):
    return name.split('_')[0]
```

```python
# Create a dictionary to store the lists from the files
mof_data = {}

# Read each file and store the data in the dictionary
for file in files:
    name = get_mof_name(file)
    mof_name = name.split('.')[0]
    file_path = os.path.join(folder_path, file)
    with open(file_path, 'r') as f:
        values = [float(line.strip()) for line in f.readlines()]
        mof_data[mof_name] = values

# Iterate through each row in the DataFrame
for index, row in df.iterrows():
    mof_name = get_mof_name(row['FrameworkName'])
    if mof_name in mof_data:
        values = mof_data[mof_name]
        for i, value in enumerate(values):
            df.at[index, f'Value_{i+1}'] = value


# Save the extended DataFrame to a new CSV file
df.to_csv('COF_total_features.csv', index=False)
```

## EXTRACT FROM INDIVIDUAL SUB EXCEL FOLDERS AND APPEND BY ROW TO TABLE

```python
import pandas as pd
import os

# Path to the grand CSV file and the folder containing the individual CSV files
grand_csv_path = '/content/CRAFTED_COF_geometric.csv'
folder_path = '/content/drive/MyDrive/UoM Dissertation/COF_isotherm'

# Load the grand CSV file
grand_df = pd.read_csv(grand_csv_path)

# Iterate over files in the folder
for file_name in os.listdir(folder_path):
    if file_name.endswith("UFF_CO2_298.csv"):
        # Extract framework name from the file name
        framework_name = file_name.split('_')[1]

        # Check if the framework name exists in the grand CSV file
        if framework_name in grand_df['FrameworkName'].values:
            # Load the individual CSV file
            file_path = os.path.join(folder_path, file_name)
            df = pd.read_csv(file_path)

            # Convert pressure values to column headers
            pressure_headers = df.iloc[:, 0].astype(str).tolist()
            mean_vol_values = df.iloc[:, 1].tolist()

            # Create a dictionary to hold the new columns data
            new_columns_data = dict(zip(pressure_headers, mean_vol_values))

            # Find the index of the framework in the grand DataFrame
            framework_index = grand_df[grand_df['FrameworkName'] == framework_name].index[0]

            # Add new columns to the grand DataFrame
            for pressure, mean_vol in new_columns_data.items():
                grand_df.at[framework_index, pressure] = mean_vol

# Save the updated grand CSV file
grand_df.to_csv("COF_CRAFTED_uptake_geometric.csv", index=False)

```

## CREATE FILES IN SUB FOLDERS

```python
import os
import shutil


# Paths to the main folder containing CIF files and the destination folder
main_folder = '/content/drive/MyDrive/UoM Dissertation/COF_PACMOF'
destination_folder = '/content/drive/MyDrive/UoM Dissertation/COF_CRAFTED'

# Paths to lj.csv and surface_features.py
lj_csv_path = '/content/LJ.csv'
surface_features_path = '/content/CO2_surface_features.py'
rotation = '/content/drive/MyDrive/UoM Dissertation/rotations_around_grid.json'

# Iterate over CIF files in the main folder
for cif_file in os.listdir(main_folder):
    if cif_file.endswith('.cif'):
        # Extract CIF file name without extension
        cif_name = os.path.splitext(cif_file)[0]

        # Create folder with CIF file name in destination folder
        folder_path = os.path.join(destination_folder, cif_name)
        os.makedirs(folder_path, exist_ok=True)

        # Copy CIF file to folder and rename to 1.cif
        cif_source_path = os.path.join(main_folder, cif_file)
        cif_dest_path = os.path.join(folder_path, "1.cif")
        shutil.copyfile(cif_source_path, cif_dest_path)

        # Copy lj.csv to folder
        lj_dest_path = os.path.join(folder_path, "LJ.csv")
        shutil.copyfile(lj_csv_path, lj_dest_path)

        # Copy surface_features.py to folder
        surface_features_dest_path = os.path.join(folder_path, "CO2_surface_features.py")
        shutil.copyfile(surface_features_path, surface_features_dest_path)


        # Copy rotation.py to folder
        rotation_dest_path = os.path.join(folder_path, 'rotation.json')
        shutil.copyfile(rotation, rotation_dest_path)

        print(f"Created folder '{cif_name}' and populated it with files.")


```

## ADD SPACES TO END OF LIST

```python
# Python script to add a space at the end of each line in a text file

input_file = '/content/COF667_with_space.txt'
output_file = '/content/COF_with_space.txt'

with open(input_file, 'r') as file:
    lines = file.readlines()

with open(output_file, 'w') as file:
    for line in lines:
        file.write(line.rstrip('\n') + ' \n')
```

## COLLLECT FILES FROM SUBFOLDERS AND RENAME

```python
import os
import shutil

# Path to the main directory containing subfolders
main_dir = '/content/drive/MyDrive/UoM Dissertation/COF_CRAFTED'
# Path to the directory where renamed files will be copied
output_dir = '/content/drive/MyDrive/UoM Dissertation/COF_SURFACE'

# Ensure the output directory exists
os.makedirs(output_dir, exist_ok=True)

# Iterate over each subfolder in the main directory
for subfolder in os.listdir(main_dir):
    subfolder_path = os.path.join(main_dir, subfolder)

    if os.path.isdir(subfolder_path):
        features_file = os.path.join(subfolder_path, 'features.txt')

        if os.path.isfile(features_file):
            new_name = f"{subfolder}.txt"
            new_path = os.path.join(output_dir, new_name)

            # Copy and rename the features.txt file to the output directory
            shutil.copy(features_file, new_path)
            print(f"Copied and renamed {features_file} to {new_path}")
```
